### copying arrays

```python
arr1=np.arange(20)
print(arr1)
```

```python
#no copy made
arr2=arr1
arr2 is arr1
```

```python
#changes in shape of one reflects in the other
arr2.shape=(2,10)
print("arr2:")
print(arr2)
print(arr1.shape)
print("arr1:")
print(arr1)
```

```python
arr2 is arr1
```

```python
#shallow coping
#changes on metadata doesn't reflect on the other
arr1=np.arange(20)
arr2=arr1[:]
arr1.shape=(4,5)
print(arr1)
print(arr2)
```

```python
#but changes on data reflects data in the other,as underlying data is the same memory
arr2[6]=100
print(arr1)
print(arr2)
```

```python
#slices are just the views pointing to the same mem
subarr=arr2[3:16]
print(subarr)
```

```python
subarr[:]=0
```

```python
print(arr1)
print(arr2)
print(subarr)
```

```python
#different ways to check if two arrays are same,views or contain their own data
print(subarr is arr1)
print(subarr is arr2)
print(arr1 is arr2)
print(subarr.base is arr1)
print(subarr.flags.owndata)
print(arr1.flags.owndata)
print(arr2.flags.owndata)
```

```python
#deep copying
coparr=arr1.copy()
coparr[:,:]=7
print(arr1)
print(coparr)
```

```python
coparr is arr1
```

```python
coparr.base is arr1
```

### Avoiding loops

```python
xmin=ymin=0.0
xmax=ymax=1.0
x=np.linspace(xmin,xmax,32,endpoint=True)
y=np.linspace(ymin,ymax,64,endpoint=True)
print(x.shape)
print(y.shape)
print(x)
print(y)
```

```python
#time elapsed
import time
M=32
N=64
x=np.linspace(0.0,1.0,M,endpoint=False)
y=np.linspace(0.0,1.0,N,endpoint=False)
```

```python
tic=time.time()
mat=np.zeros((M,N))
for i in range(M):
    for j in range(N):
        mat[i,j]=np.sin(2.0*np.pi*x[i]*y[i])
toc=time.time()
print("time elapsed:{} sec".format(toc-tic))
mat
```

```python
#extend 1d arrays to 2d
x2,y2=np.meshgrid(x,y,indexing='ij')
print(x2)
print(x2[:,0])
print(x2[0,:])
```

```python
#time elapsed using the above for the same operation(gets reduced)
print(mat)
tic=time.time()
mat=np.sin(2.0*np.pi*x2*y2)
mat
toc=time.time()
print("time elapsed:{} sec".format(toc-tic))
```

### Splitting of arrays

``np.hsplit`` ``np.vsplit`` ``np.split``

```python
chain=[1,2,3,4,1,2,3,4,6,5,43,8,7,6]
c1,c2,c3,c4=np.split(chain,[4,7,9])
print(c1,c2,c3,c4)
```

```python
child=np.arange(16).reshape((4,4))
child
```

```python
mom,dad=np.hsplit(child,[2])
print(mom)
print(dad)
```

```python
mom,dad=np.vsplit(child,[2])
print(mom)
print(dad)
```
