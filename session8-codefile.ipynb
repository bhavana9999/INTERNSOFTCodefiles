## Visualizing data using matplotlib

```python
#importing matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#plot() to plot the datapoints
plt.plot(np.random.normal(size=100),np.random.normal(size=100),'dr')
```

```python
%matplotlib inline
```

```python
plt.plot(np.random.normal(size=400), np.random.normal(size=400), 'o')
```

```python
#plotting in pandas
normals=pd.Series(np.random.normal(size=50))
print(normals)
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
#plotting for a dataframe
accu=pd.DataFrame({'normal':np.random.normal(size=10),'gamma': np.random.gamma(1, size=10),'poisson': np.random.poisson(size=10)})
print(accu)
accu.cumsum(0).plot()
```

```python
#subplots
accu.cumsum(0).plot(subplots=True)
```

```python
#using secondary y-axis for normal 
accu.cumsum(0).plot(secondary_y='normal')
```

#### matplotlib's subplots

```python
fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(6,18))
for i,var in enumerate(['normal','gamma','poisson']):
    accu[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

### Simple line plots

```python
figure1=plt.figure()
ax=plt.axes()
deg=np.linspace(0,40,100)
print(deg)
ax.plot(deg,np.sin(deg))
```

```python
plt.plot(deg,np.sin(deg))
```

### Line colors and style

```python
#colors
plt.plot(deg,np.sin(deg-0),color='green')
plt.plot(deg,np.sin(deg-1),color='y')
plt.plot(deg,np.sin(deg-2),color='0.65')
```

```python
#linestyles
plt.plot(deg,np.sin(deg-0),linestyle='-')
plt.plot(deg,np.sin(deg-1),linestyle='-.')
plt.plot(deg,np.sin(deg-2),linestyle=':')
```

### Adjusting the plot:Axes limits

```python
#limits for x and y axes of the plot
plt.plot(deg,np.sin(deg-1))
plt.xlim(-1,41)
plt.ylim(-1.5,1.5)
```

### labelling plots

```python
#setting titles,xlabel,ylabel for the plot
plt.plot(deg,np.cos(deg))
plt.title("cosine curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

### Scatter plots

```python
x=np.linspace(2,20,40)
y=np.log(x)
plt.plot(x,y,'o',color='red')
```

```python
plt.plot(x,y,'-ob')
```

### plt.scatter for scatter plots

```python
plt.scatter(x,y,marker='o')
```

```python
pattern= np.random.RandomState(5)
x = pattern.randn(100)
y =pattern.randn(100)
col=pattern.rand(100)
S= 1000*pattern.rand(100)
plt.scatter(x, y, c=col, s=S, alpha=0.8,cmap='Blues')
plt.colorbar()
```

### Visualizing errors

#### errorbars

```python
#yerror
x = np.linspace(0,40, 100)
dy = 0.2
y = np.sin(x)+dy*np.random.randn(100)
plt.errorbar(x, y,yerr=dy,fmt='.r')
```

### Histograms

```python
plt.style.use('seaborn-white')
d1= np.random.randn(10000)
print(d1)
```

```python
plt.hist(d1)
```

```python
plt.hist(d1, bins=8, density=True, alpha=0.6,histtype='stepfilled')
```

```python
x = np.random.normal(1,2.8, 1000)
y = np.random.normal(-2,2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.5, density=True, bins=50)
plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

```python
counts, bin_edges = np.histogram(d1, bins=8)
print(counts)
```

### Two-dimensional histograms

```python
mean = [5, 2]
covariance = [[3, 1], [2, 0.5]]
x, y = np.random.multivariate_normal(mean,covariance,10000).T
```

```python
plt.hist2d(x, y, bins=20, cmap='Reds')
cb = plt.colorbar()
```

```python
#plt.hexbin for hexagonal binning
plt.hexbin(x,y,gridsize=40,cmap='Blues')
cb = plt.colorbar()
```
