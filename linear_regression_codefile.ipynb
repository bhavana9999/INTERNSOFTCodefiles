### Linear Regression

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("advertising.csv")
```

```python
data.head()
```

```python
#shape
data.shape
```

```python
#visualizing the relationship between the dependent and independent attributes using scatterplots
fig,axes=plt.subplots(1,3,sharey=True)
data.plot(kind="scatter",x='TV',y='Sales',ax=axes[0],figsize=(20,10))
data.plot(kind="scatter",x='Newspaper',y='Sales',ax=axes[1])
data.plot(kind="scatter",x='Radio',y='Sales',ax=axes[2])        
```

### Scikit-learn

```python
#create x and y
feature_cols=['TV']
X=data[feature_cols]
y=data.Sales

#follow the usual sklearn pattern,import,instantiate,fit
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)

#print intercept and coefficients
print(lr.intercept_)
print(lr.coef_)
```

### plotting the least squares ljne

```python
#creating a dataframe since the statsmodel formula interface xpects it
X_new=pd.DataFrame({'TV':[data.TV.min(),data.TV.max()]})
X_new.head()
```

```python
preds=lr.predict(X_new)
preds
```

```python
#plotting the observed data
data.plot(kind='scatter',x='TV',y='Sales')
#plottig the least squared line
plt.plot(X_new,preds,c='red',linewidth=4)
```

### confidence in our model

```python
import statsmodels.formula.api as smf
lm=smf.ols(formula='Sales ~ TV',data=data).fit()
lm.conf_int()
```

```python
#printing the p-values for the model coefficients
lm.pvalues
```

```python
# printing the R-squared value for the model
lm.rsquared
```

### Multiple Linear Regression

```python
# create X and y
feature_cols = ['TV', 'Radio', 'Newspaper']
X = data[feature_cols]
y = data.Sales

lm = LinearRegression()
lm.fit(X, y)

# print intercept and coefficients
print(lm.intercept_)
print(lm.coef_)
```

```python
lm = smf.ols(formula='Sales ~ TV + Radio + Newspaper', data=data).fit()
lm.conf_int()
lm.summary()
```

### Feature Selection

```python
# only include TV and Radio in the model
lm = smf.ols(formula='Sales ~ TV + Radio  ', data=data).fit()
lm.rsquared
```

```python
# add Newspaper to the model (which we believe has no association with Sales)
lm = smf.ols(formula='Sales ~ TV + Radio + Newspaper', data=data).fit()
lm.rsquared
```
